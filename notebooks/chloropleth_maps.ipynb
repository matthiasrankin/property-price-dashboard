```python
import geopandas as gpd
from property_price_dashboard.plotting import plot_chloropleth
from collections import defaultdict
```

```python
region_shapefile_and_key_mapping = {
    "District Electoral Area": {
        "shapefile": "../data/chloropleth_data/electoral_areas.csv",
        "region_key": "FinalR_DEA"
    },
    "Electoral Ward": {
        "shapefile": "../data/chloropleth_data/electoral_wards.csv",
        "region_key": "WARDNAME"
    }
}

property_type_mapping = {
    "All Residential": "median_sale_price_total",
    "Detached": "median_sale_price_det",
    "Semi-Detached": "median_sale_price_sdt",
    "Terrace": "median_sale_price_ter",
    "Apartment": "median_sale_price_apt"
}
```

```python
chloropleths = defaultdict(dict)

for property_type, price_key in property_type_mapping.items():
    for region_type, region_dict in region_shapefile_and_key_mapping.items():
        region_key = region_dict["region_key"]
        shapefile_path = region_dict["shapefile"]

        gdf = gpd.read_file(shapefile_path)

        my_map = plot_chloropleth(
            gdf, region_key, price_key, region_type, f"Median Sale Price ({property_type})"
        )

        chloropleths[property_type][region_type] = my_map
```

```python
chloropleths = dict(chloropleths)
```

```python
import json

with open("../static/chloropleths.json", "w", encoding="utf-8") as file_:
    json.dump(chloropleths, file_)
```
